```python
print(data['annotation'][1][0]['imageHeight'])
print(data['annotation'][1][0]['imageWidth'])
```

```python
x_min = []
x_max = []
y_min = []
y_max = []
h = []
w = []

for i in range(237):
    x_min.append(data['annotation'][i][0]['points'][0]['x'])
    x_max.append(data['annotation'][i][0]['points'][1]['x'])
    y_min.append(data['annotation'][i][0]['points'][0]['y'])
    y_max.append(data['annotation'][i][0]['points'][1]['y'])
    h.append(data['annotation'][i][0]['imageHeight'])
    w.append(data['annotation'][i][0]['imageWidth'])
```

```python
type(x_min)
```

```python
df['x_min'] = x_min
df['y_min'] = y_min
df['x_max'] = x_max
df['y_max'] = y_max
```

```python
df['width'] = w
df['height'] = h
```

```python
df['class'] = ['license']*237
```

```python
df['filename'] = [str(i)+'jpg' for i in range(237)]
```

```python
df.head()
```

```python
df.drop(['content','annotation'],axis=1,inplace=True)
```
